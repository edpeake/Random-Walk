```python
# Python code for 3D random walk, These are biased to move in a particular way

from __future__ import division
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import FuncAnimation, PillowWriter

import matplotlib.pyplot as plt 
import numpy 
import random 

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
  
# defining the number of steps 
n = 20000
```

## First Random Walk

These are biased to move in a particular direction, in this case by the addition of 0.01 to the +z direction. 

```python
#creating two array for containing x and y coordinate 
#of size equals to the number of size and filled up with 0's 

x = 100+numpy.zeros(n) 
y = numpy.zeros(n) 
z = 100+numpy.zeros(n) 

# Random walk Number 1 
for i in range(1, n): 
    val = random.randint(1, 6) 
    if val == 1: 
        x[i] = x[i - 1] + 1
        y[i] = y[i - 1] 
        z[i] = z[i - 1] 
    elif val == 2: 
        x[i] = x[i - 1] - 1
        y[i] = y[i - 1]
        z[i] = z[i - 1]
    elif val == 3: 
        x[i] = x[i - 1] 
        y[i] = y[i - 1] + 1
        z[i] = z[i - 1]
    elif val == 4: 
        x[i] = x[i - 1] 
        y[i] = y[i - 1] 
        z[i] = z[i - 1] + 1.01
    elif val == 5: 
        x[i] = x[i - 1] 
        y[i] = y[i - 1] 
        z[i] = z[i - 1] - 1
    else: 
        x[i] = x[i - 1] 
        y[i] = y[i - 1] - 1
        z[i] = z[i - 1]


fig = plt.figure()

ax = fig.add_subplot(111, projection='3d')
ax.grid(False)

ax.scatter3D(x, y, z,c='blue',alpha=1,s=0.1);

plt.axis('off')
plt.savefig('fig1.png', dpi = 300)
```

## Second Random Walk

```python
x1 = numpy.zeros(n) 
y1 = numpy.zeros(n) 
z1 = -100+numpy.zeros(n) 
        
# Random walk Number 2
        
for i in range(1, n): 
    val = random.randint(1, 6) 
    if val == 1: 
        x1[i] = x1[i - 1] + 1.01
        y1[i] = y1[i - 1] 
        z1[i] = z1[i - 1]
    elif val == 2: 
        x1[i] = x1[i - 1] - 1
        y1[i] = y1[i - 1]
        z1[i] = z1[i - 1]
    elif val == 3: 
        x1[i] = x1[i - 1] 
        y1[i] = y1[i - 1] + 1.01
        z1[i] = z1[i - 1]
    elif val == 4: 
        x1[i] = x1[i - 1] 
        y1[i] = y1[i - 1] 
        z1[i] = z1[i - 1] + 1.01
    elif val == 5: 
        x1[i] = x1[i - 1] 
        y1[i] = y1[i - 1] 
        z1[i] = z1[i - 1] - 1
    else: 
        x1[i] = x1[i - 1] 
        y1[i] = y1[i - 1] - 1
        z1[i] = z1[i - 1]

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.grid(False)
ax.xaxis.pane.fill = ax.yaxis.pane.fill = ax.zaxis.pane.fill = False
ax.scatter3D(x1, y1, z1,c='red',alpha=1,s=0.1);

plt.axis('off')
plt.savefig('fig2.png', dpi = 300)
```

## Thrid Random Walk

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x2 = numpy.zeros(n) 
y2 = numpy.zeros(n) 
z2 = numpy.zeros(n)     
        
# Random walk Number 3      

for i in range(1, n): 
    val = random.randint(1, 6) 
    if val == 1: 
        x2[i] = x2[i - 1] + 1
        y2[i] = y2[i - 1] 
        z2[i] = z2[i - 1]
    elif val == 2: 
        x2[i] = x2[i - 1] - 1.01
        y2[i] = y2[i - 1]
        z2[i] = z2[i - 1]
    elif val == 3: 
        x2[i] = x2[i - 1] 
        y2[i] = y2[i - 1] + 1
        z2[i] = z2[i - 1]
    elif val == 4: 
        x2[i] = x2[i - 1] 
        y2[i] = y2[i - 1] 
        z2[i] = z2[i - 1] + 1
    elif val == 5: 
        x2[i] = x2[i - 1] 
        y2[i] = y2[i - 1] 
        z2[i] = z2[i - 1] - 1.01
    else: 
        x2[i] = x2[i - 1] 
        y2[i] = y2[i - 1] - 1
        z2[i] = z2[i - 1]

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.grid(False)
ax.xaxis.pane.fill = ax.yaxis.pane.fill = ax.zaxis.pane.fill = False
ax.scatter3D(x2, y2, z2,c='green',alpha=1,s=0.1);

plt.axis('off')
plt.savefig('fig3.png', dpi = 300)
```
